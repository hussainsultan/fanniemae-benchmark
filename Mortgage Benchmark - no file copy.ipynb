```python
def create_db(datadir):
    global env
    print(f"{inspect.stack()[0][3]}  entered")
    perf_files_list = map(lambda x : "'" + env["perf_path"] + env["path_separator"] + x +"'", env["perf_list"])
    env["perf_files"] = ",".join(perf_files_list)
    acq_files_list = map(lambda x : "'" + env["acq_path"] + env["path_separator"] + x +"'", env["acq_list"])
    env['acq_files'] = ",".join(acq_files_list)
    conn = duckdb.connect("mortgage.db")
    conn.execute(f"CREATE OR REPLACE VIEW perf AS SELECT * FROM {env['perf_files']}")
    conn.execute(f"CREATE OR REPLACE VIEW acq AS SELECT * FROM {env['acq_files']}")
    conn.close()
```

```python
def summary_expr(datadir):
    print(f"{inspect.stack()[0][3]}  entered")
    db = ibis.duckdb.connect("mortgage.db")
    perf = db.table("perf")
    acq = db.table("acq")
    t0 = time.time()
    acq = acq[
        acq.loan_id, acq.orig_date.split("/")[1].name("year"), acq.borrower_credit_score
    ]
    joined = acq.inner_join(perf, acq.loan_id == perf.loan_id)
    t1 = time.time()
    chargeoffs = (
        ibis.case()
        .when(
            (perf.zero_balance_code.isin(["02", "03", "09", "15"]))
            & (perf.disposition_date.notnull()),
            1,
        )
        .else_(0)
        .end()
        .name("charegoffs")
    )
    t2 = time.time()
    
    dollar_co = (
        perf.zero_balance_code.isin(["02", "03", "09", "15"])
        & (perf.disposition_date.notnull())
    ).ifelse(perf.current_actual_upb, 0)
    t3 = time.time()
    
    loans = joined.mutate(chargeoffs=chargeoffs, dollar_co=dollar_co).projection(
        [
            perf.loan_id,
            chargeoffs,
            dollar_co.name("dollar_co"),
            perf.loan_age,
            perf.current_actual_upb,
            acq.year,
            acq.borrower_credit_score,
        ]
    )
    t4 = time.time()
    
    summary = (
        loans[loans.loan_age > 0]
        .groupby([loans.year, loans.loan_age])
        .aggregate(
            co_count=lambda x: x.charegoffs.cast("int64").sum(),
            dollar_co=lambda x: x.dollar_co.sum(),
            avg_credit_score=lambda x: x.borrower_credit_score.mean(),
            upb_sum=lambda x: x.current_actual_upb.sum(),
        )
    )
    t5 = time.time()
    
    acq_agg = acq.groupby([acq.year]).loan_id.count()
    t6 = time.time()
    
    summary = summary.inner_join(acq_agg, acq_agg.year == summary.year)
    t7 = time.time()
    
    summary = summary.projection(
        [
            summary.year_x,
            summary.loan_age,
            summary["count(loan_id)"],
            summary.avg_credit_score,
            summary.upb_sum,
            summary.dollar_co,
        ]
    )
    t8 = time.time()
    print(f"t1: {t1-t0}\t t2: {t2-t1}\t t3: {t3-t2}\t t4: {t4-t3}\t t5: {t5-t4}\t t6: {t6-t5}\t t7: {t7-t6}")
    del db
    return summary
```

```python
def summary_sql(datadir):
    global env
    print(f"{inspect.stack()[0][3]}  entered")
    with open("summary.sql") as f:
        template = Template(f.read())
    return template.render(
        perf=str(env['perf_files']), acq=str(env['acq_files'])
    )


```

```python
def window_sql(datadir):
    global env
    print(f"{inspect.stack()[0][3]}  entered")
    sql = f"select count(*) from (select RANK() OVER (PARTITION BY loan_id ORDER BY monthly_reporting_period) as number from env['perf_files'])"
    return sql



```

```python
def execute(expr, threads=8):
    print(f"{inspect.stack()[0][3]}  entered")
    conn = duckdb.connect("mortgage.db")
    print("connect to duckdb")
    print(f"PRAGMA threads={threads}")
    conn.execute(f"PRAGMA threads={threads}")
    print("conn.execute - PRAGMA complete")
    print(f"{str(expr)}")
    result = conn.execute(str(expr)).fetchall()
    print("conn.excute str(expr) complete")
    conn.close()
    return result


```

```python
def platform_info():
    print(f"{inspect.stack()[0][3]}  entered")
    return {
        "machine": platform.machine(),
        "version": platform.version(),
        "platform": platform.platform(),
        "system": platform.system(),
        "cpu_count": psutil.cpu_count(),
        "memory": psutil.virtual_memory().total,
        "processor": platform.processor(),
    }
```

```python
def main(mode, datadir, threads):
    global thread_count
    print(f"{inspect.stack()[0][3]}  entered")
    datadir = Path(datadir)
    # t3 = time.time()
    create_db(datadir)
    # t4 = time.time()
    # print(f"time to create database: {t4-t3}")
    
    if mode == "ibis":
        summary = summary_expr(datadir)
        sql = summary.compile().compile(compile_kwargs={"literal_binds": True})
    else:
        sql = summary_sql(datadir)
    #t5 = time.time()
    row_count_perf = execute("select count(*) from perf",env["thread_count"])[0][0]
    row_count_acq = execute("select count(*) from acq",env["thread_count"])[0][0]
    #t6 = time.time()
    #print(f"time to count rows: {t6-t5}")
    #if threads:
        #runs = range(2, psutil.cpu_count() + 2, 2)
    #else:
        #runs = [psutil.cpu_count()]
    runs = [env["thread_count"]]
    result = []
    for thread in runs:
        start_time = time.time()
        print(f"threads: {thread} start: {start_time}")
        mem = memory_usage(
            (
                execute,
                (
                    sql,
                    thread,
                ),
            )
        )
        total_time = time.time() - start_time

        data = {
            **platform_info(),
            "threads": thread,
            "run_date": datetime.datetime.now().strftime("%d/%m/%Y %H:%M:%S"),
            "total_time": total_time,
            "row_count_perf": row_count_perf,
            "row_count_acq": row_count_acq,
            "max_memory_usage": max(mem),
            "incremental_memory_usage": mem[-1] - mem[0],
            "sql": " ".join(str(sql).split()),
        }
        result.append(data)
    # print(json.dumps(result))

    return(result)
```

```python
def threadCompare():
    print(f"{inspect.stack()[0][3]}  entered")
    t1 = []
    _cpus = (platform_info())["cpu_count"]
    for y in range (2, 17, 2):

        for t in range(1, _cpus+1):
            setThreadsAndYears(t,y)
            setGlobals()
            copyMortgageFiles()

            res  = main(env["mode"], env["data_folder"], env["thread_count"])

            print(f"CPUs: {res[0]['cpu_count']} Threads: {res[0]['threads']} \t Years: {env['year_count']} \t Elapsed Time: {res[0]['total_time']} ")
            res[0]['year_count'] = env['year_count']
            t1.append(res[0])
    return t1
```

```python
def getVariance():
    # change the _threads tuple to represent the range of threads to use for this run
    # change the _years tuple to represent the year sets to execute against. 
    # year range can be from 1 to 17.
    # set iterations to be the number of times each query is run. Default is 10.
    # iterations is used to determin average and standard deviation
    print(f"{inspect.stack()[0][3]}  entered")
    _threads = range(2,8)
    _years = range(2,13,2)
    _iterations = 10
    _res = []
    _iter = 0
    for y in _years:
        for t in _threads:
            for i in range(0,_iterations):
                setThreadsAndYears(t,y)
                setGlobals()
                copyMortgageFiles()
                res  = main(env["mode"], env["data_folder"], env["thread_count"])
                _res.append({"iteration": i, "threads": res[0]['cpu_count'], "years": env['year_count'], "e_time": res[0]['total_time']})
                _iter +=1
                print(f"[{i}]\tCPUs: {res[0]['cpu_count']} Threads: {res[0]['threads']} \t Years: {env['year_count']} \t Elapsed Time: {res[0]['total_time']} ")
                res[0]['year_count'] = env['year_count']
    return _res
```
